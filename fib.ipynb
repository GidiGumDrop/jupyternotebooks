```python
from math import sqrt
import matplotlib.pyplot as plt
```

```python
start = -5
end = 5 # not inclusive by 0.1

phi = (1+sqrt(5))/2
```

```python
def fib(n):
    return ((phi**n) - ((-phi)**-n))/sqrt(5)
```

```python
x = []
y = []

# increments of 0.1
values = [i/10 for i in range(start*10,end*10)]

for i in values:
    num = fib(i)
    x.append(num.real)
    y.append(num.imag)
```

```python
plt.axhline(y=0, color='black', linestyle='-')
plt.plot(x,y)
plt.xlabel('real')
plt.ylabel('imaginary')
plt.title(f'interval: {0.1}, start: {start}, end: {end}')
plt.show()
```
